```python
def get_entries_count(soup):
    item_link_selector = "#site-content > div:nth-child(2) > div.sc-AngWr.jxhegR > div > div > ul"
    entry = soup.select(
        item_link_selector
    )[0]
    count = len([*entry.children])
    return count 

def get_entry(soup, result_id):

    #name
    item_link_selector = lambda x : f'#site-content > div:nth-child(2) > div.sc-AngWr.jxhegR > div > div > ul > li:nth-child({x}) > div.sc-drMgrp.gxpLdE > a > div > div.sc-FjMCv.gwKDa-D > div'
    print(item_link_selector(result_id))
    entry = soup.select(
        item_link_selector(result_id)
    )[0]
    title = entry.text

    #rating
    item_link_selector = lambda x : f'#site-content > div:nth-child(2) > div.sc-AngWr.jxhegR > div > div > ul > li:nth-child({x}) > div.sc-drMgrp.gxpLdE > div > div > span'

    entry = soup.select(
        item_link_selector(result_id)
    )[0]
    rating = entry.text

    #medal
    item_link_selector = lambda x : f'#site-content > div:nth-child(2) > div.sc-AngWr.jxhegR > div > div > ul > li:nth-child({x}) > div.sc-drMgrp.gxpLdE > div > span > span'
    entries = soup.select(
        item_link_selector(result_id)
    )

    medal = "No medal"
    if len(entries) > 0:
        entry = entries[0]
        medal = entry.text.strip()

    #link
    item_link_selector = lambda x : f'#site-content > div:nth-child(2) > div.sc-AngWr.jxhegR > div > div > ul > li:nth-child({x}) > div.sc-drMgrp.gxpLdE > a > div > div.sc-FjMCv.gwKDa-D > span:nth-child(3) > span > a'

    entry = soup.select(
        item_link_selector(result_id)
    )[0]

    link_suffix = entry["href"][:-9]
    link = f"https://www.kaggle.com{link_suffix}"

    #source
    item_link_selector = lambda x : f'#site-content > div:nth-child(2) > div.sc-AngWr.jxhegR > div > div > ul > li:nth-child({x}) > div.sc-drMgrp.gxpLdE > a > div > div.sc-FjMCv.gwKDa-D > span:nth-child(3) > span > span.sc-jtCqdw.huegwV > span'
    entries = soup.select(
        item_link_selector(result_id)
    )

    source = "No source"
    if len(entries) > 0:
        entry = entries[0]
        source = entry.text.strip()

    return {
        "title": title,
        "rating": rating,
        "medal": medal,
        "link": link,
        "source": source
    }

entries = []
def parse_code_notebooks(url):
    chromedriver_path = "chromedriver.exe"


    service = Service(executable_path=chromedriver_path)
    options = webdriver.ChromeOptions()
    options.add_argument('--headless')
    driver = webdriver.Chrome(service=service, options=options)

    results_found = True
    failed_once = False
    i = 1
    while results_found and i < 500:
        #try:
            driver.get(f"{url}{i}")

            #element = WebDriverWait(driver, 10).until(
            #    EC.presence_of_element_located((By.CSS_SELECTOR, "#site-content > div:nth-child(2) > div.sc-AngWr.jxhegR > div > div > ul > li:nth-child(1) > div.sc-drMgrp.gxpLdE > div"))
            #)

            html = driver.page_source
            soup = BeautifulSoup(html, 'html.parser')
            entries_count = get_entries_count(soup)

            if entries_count > 0:
                failed_once = False
                entries.extend(
                    [get_entry(soup, j) for j in range(1, entries_count+1)]
                )
                i += 1
            else:
                results_found = False
                
            print(f"parsed page {i} parsed totalling in {len(entries)} entries")

        # except Exception as inst:
        #     if failed_once:
        #         results_found = False
        #     failed_once = True
        #     i += 1
        #     print(f"parsing failed on page {i}")
        
            time.sleep(2)

    print(f"parsing ended with {i} pages parsed totalling in {len(entries)} entries")

    return entries
```
